```python
import os
import pandas as pd
import numpy as np
import random
```

```python
county_data_name = 'Assessor_Parcels_Data_-_2006_thru_2019' # we're going to start with one file
file_path = r'C:\Users\kas\Downloads\Assessor_Parcels_Data_-_2006_thru_2019.csv'
file_name = 'Assessor_Parcels_Data_-_2006_thru_2019.csv'
_2006_2019_url = 'https://data.lacounty.gov/api/views/9trm-uz8i/rows.csv?accessType=DOWNLOAD'
```

getting files frome folder 'complete' that end with .xlsx:

```python
path = os.getcwd()
files = os.listdir(path)
files = [f for f in files if f[-4:] == 'xlsx']
files[:5]
```

making a list of all the APNs in the files from folder 'complete':

```python
ids = []
bad_files = []
col_name = 'Assessor\'s Parcel Number (APN)  '
for f in files:
    tmp = pd.read_excel(f, header = 5)
    if not (col_name in tmp.columns) or not (tmp.columns[0] == col_name):
        bad_files.append(f)
    else:
        tmp = tmp[col_name]
        ids.extend(tmp)
print('bad files ',len(bad_files), ', total ID\'s ', len(ids))
```

```python
bad_files2 = []
col_name = 'Assessor\'s Parcel Number (APN)'  # there was a space at the end that these guys took out
for f in bad_files:
    tmp = pd.read_excel(f, header = 5)
    if not (col_name in tmp.columns) or not (tmp.columns[0] == col_name):
        bad_files2.append(f)
    else:
        tmp = tmp[col_name]
        ids.extend(tmp)
print('bad files ',len(bad_files2), ', total ID\'s ', len(ids))
```

```python
bad_files3 = []
col_name = 'Assessor\'s Parcel Number (APN)'  # header == 8
for f in bad_files2:
    tmp = pd.read_excel(f, header = 8)
    if not (col_name in tmp.columns) or not (tmp.columns[0] == col_name):
        bad_files3.append(f)
    else:
        tmp = tmp[col_name]
        ids.extend(tmp)
print('bad files ',len(bad_files3), ', total ID\'s ', len(ids))
```

reading county data:

```python
pdata = pd.read_csv(file_path, engine = 'python')
pdata.head()
```

and we're done

```python
filtered = pdata[pdata.AssessorID.isin(ids)]
filtered.head()
```

```python
out_file_name = 'filtered_' + file_name
filtered.to_csv(out_file_name, index = False)
```
